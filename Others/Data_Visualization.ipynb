## (Scatter Plots)When column is category and index is record of individual ones

- sns.scatterplot()
- sns.regplot() # adds a regression trend line to scatterplot()
- sns.lmplot() # used with hue and adds two regeession line to scatterplot()
- sns.swarmplot()

```python
df = pd.read_csv('/content/Dataset/insurance.csv')
df.head()
```

```python
sns.scatterplot(x=df['bmi'], y=df['charges'])
```

If you use the 'hue' parameter in sns.scatterplot(), it will assign different colors to data points based on the categories of the specified column. It's important to note that the column used for 'hue' should be categorical.

```python
sns.scatterplot(x=df['bmi'], y=df['charges'], hue=df['smoker'])
```

The difference between sns.regplot() and sns.scatterplot() is that sns.regplot() adds a regression line representing the trend, while sns.scatterplot() only plots the data points without a trend line.

```python
sns.regplot(x=df['bmi'], y=df['charges'])
```

sns.lmplot() is part of Seaborn's earlier API, which uses a slightly different syntax. The key difference between sns.scatterplot() and sns.lmplot() is that the latter automatically adds a regression trend line to the scatter plot.

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=df)
```

```python
sns.swarmplot(x=df['smoker'], y=df['charges'])
```

## (Distributions)When column is category and index is record of individual ones

- sns.histplot() # distribution in bar graph, # also can be used with hue
- sns.kdeplot() # distribution in kde
- sns.joinplot() # show distribution of two in kde


```python
df = pd.read_csv('/content/Dataset/iris.csv', index_col='Id')
df.head()
```

```python
sns.histplot(df['Petal Length (cm)'])
```

```python
sns.histplot(data=df, x='Petal Length (cm)', hue='Species')
plt.title("some title")
```

```python
sns.kdeplot(data=df['Petal Length (cm)'], shade=True)
```

```python
sns.jointplot(x=df['Petal Length (cm)'], y=df['Sepal Width (cm)'], kind="kde")
```
