## Creating the Dataset

### DataFrame

Here we pass a dictionary to `pd.DataFrame()`. The key of the dictionary is name of the column and values are cllumn values.

```python
pd.DataFrame({'Yes': [59, 45, 57], 'No': [12, 13, 7]})
```

By defult, the index of df increment from 0 to numOfRows-1, but you can set it to have own custom index.

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'],
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

### Series

You can think of Series as a single column from DataFrame. Here you pass a list to` pd.Series()`.

```python
pd.Series([1,2,3,4,5])
```

You can also provide your own custom index and name of the Series

```python
pd.Series([30, 35, 40], index=["2015 Sales", "2016 Sales", "2017 Sales"], name='Product A')
```

Here first value also represent row and second value represent column. First value are alike integer, and second value are alike String.

```python
df.loc[0, 'country']
```

```python
df.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

The key difference between df.iloc and df.loc is this:

df.iloc use positional index and df.loc use index label.

So for example let's say we have a DaatFrame where index starts from 345.
df.iloc[10] will gives us 354th row, but if you do df.loc[10] it will throw an error saying out of range.

## Modify

### Group

Here df.groupby('points') is grouping given DataFrame by points. So as there are 21 unique points there will be total 21 subgroups.

Then for each sub groups we're accessing its price column and then getting mean.


```python
df.groupby('points').price.mean()
```

In following code, we are creating subgoups by winery. Then for each subgroup `DataFrame`, we only get sub `DataFrame` which only have title column, the getting the first row by `.iloc[0]`.

```python
df.groupby('winery').apply(lambda df: df.title.iloc[0])
```

df.points.idxmax() returns index label.

```python
df.points.idxmax()
```

```python
df.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
df.groupby(['country']).price.agg([min, max])
```

### Combining

When combing both DaatFrame should have same columns

**Option1 - use pd.concat()**

```python
canadian_youtube = pd.read_csv("../input/youtube-new/CAvideos.csv")
british_youtube = pd.read_csv("../input/youtube-new/GBvideos.csv")

pd.concat([canadian_youtube, british_youtube])
```

**Option2 - use pd.join()**

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```
